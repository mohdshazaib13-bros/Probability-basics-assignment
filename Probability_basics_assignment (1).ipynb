3.  A bag has 3 red and 2 blue balls. If one ball is drawn randomly and is red, what is the probability
that the next ball is also red (without replacement)?

    - The probability that the next ball is also red is = 0.5.


4. The population of a school is divided into 60% boys and 40% girls. If you want equal representation
of both genders in the sample, which method should you use: Simple Random Sampling or Stratified Sampling?
Why?

   - The population is unevenly divided (60% boys, 40% girls),To ensure equal representation of both genders in the sample, the population should be divided into strata (boys and girls),Then, equal numbers are selected from each stratum.


5.  The average height of 1000 students = 160 cm. A sample of 100 students shows an average height=158cm. Find the sampling error.

   - The sampling error is −2 cm, meaning the sample average is 2 cm less than the population average.

6.  The population mean salary is ₹50,000 with σ = ₹5,000. If we take a sample of 100 employees, what
is the standard error of the mean (SEM)?

   - The standard error of the mean (SEM) is ₹500.

10.  Differentiate between discrete and continuous random variables with examples.

   - Discrete - Takes countable values (finite or countably infinite),Usually whole numbers,Obtained by counting,Probability is assigned to specific values.

   - Continuous - Takes uncountable values within a range,Can take any real value (including decimals),Obtained by measuring,Probability is assigned to an interval.
